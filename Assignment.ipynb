```python
#imports
import os
from os.path import dirname as up, join

from urllib.parse import unquote, urlparse
from pathlib import PurePosixPath

from datetime import datetime

import requests
import json

import sqlite3

import pandas as pd


#Function to get secrets like SQL queries and Keys stored in JSON files
def get_secrets(type, name):
    cwd = os.getcwd()
    
    file_path = join(cwd, "secrets", type + ".json") 
    
    with open(file_path, "r") as file:
        data = json.load(file)
    
    return data[name]

#Function to perform DQL queries on the local SQLite DB
def query_database(owner, repo, query_name):
        
        conn = sqlite3.connect('github_commits.db')
    
        query = get_secrets("DQL", query_name).format(**{"owner": owner, "repo": repo})
        
        df = pd.read_sql_query(query, conn)
        
        conn.close()
        
        return df
    
#Recursively calls the Github API for commits and stores them in the local SQLite DB 
def etl():
    
    #Create DB and Helper tables
    def prerequisites(owner, repo):
        
        create_database()
        
        create_table(owner, repo)
        
        create_weekday_table()
        
        insert_into_weekday_table()
        
        create_buckets_table()
        
        insert_into_buckets_table()
        
    #Ask user for the repository
    def ask_repo_url():
        def parse_url(url):
            return PurePosixPath(
                    unquote(
                        urlparse(
                            url
                        ).path
                    )
                ).parts
        
        while True:
            url = input("Enter the repository url in the format 'https://github.com/{owner}/{repo_name}' \n Ex:- 'https://github.com/apache/airflow':\n")

            try:
                parts = parse_url(url)
                if len(parts) == 3:
                    break
                else:
                    print("The format is wrong!")
            except:
                print("Couldn't parse the url!", end = ' ')
        
        return parts[1: 3]
    
    #Ask user for the start date
    def ask_start_date():
        while True:
            since = input("Enter the start date in the format(YYYY-MM-DD): ")
            
            try:
                since = datetime.strptime(since, "%Y-%m-%d")
                break
            except:
                print("The format is wrong!", end = '')
        
        return since
    
    #Ask user for the end date
    def ask_end_date():
        while True:
            until = input("Enter the end date in the format(YYYY-MM-DD)\nNote: Only commits before this date will be considered: ")
            
            try:
                until = datetime.strptime(until, "%Y-%m-%d")
                break
            except:
                print("The format is wrong!", end = '')
                
        return until
        
    #Check to see if the repository exists
    def get_commits(owner, repo, since, until):
        subs = {"owner": owner, "repo": repo}
        
        LIST_COMMITS_URL = get_secrets("API", "url").format(**subs)
        ACCESS_TOKEN = get_secrets("API", "token")
        
        headers = {
            "Accept": "application/vnd.github+json",
            "Authorization": "Bearer {bearer}".format(bearer = ACCESS_TOKEN),
            "X-GitHub-Api-Version": "2022-11-28"
        }
        
        params = {"since": since.isoformat(), "until": until.isoformat()}
        
        page1 = requests.get(LIST_COMMITS_URL, headers = headers, params = params)
        
        return page1
    
    #Get all the commits for the repository
    def get_all_commits(owner, repo, since, until):
        subs = {"owner": owner, "repo": repo}
        
        #url to call
        LIST_COMMITS_URL = get_secrets("API", "url").format(**subs)
        #Personal access token from Github for higher rates
        ACCESS_TOKEN = get_secrets("API", "token")
        
        headers = {
            "Accept": "application/vnd.github+json",
            "Authorization": "Bearer {bearer}".format(bearer = ACCESS_TOKEN),
            "X-GitHub-Api-Version": "2022-11-28"
        }
        
        #Limit commits to the date range
        params = {"since": since.isoformat(), "until": until.isoformat(), "per_page": 100}
        
        data = list()
        url = LIST_COMMITS_URL
        #Loop through and load data into DB. DO until next url doesn't exist
        while url :
            response = requests.get(url, headers = headers, params = params)
            
            #Check if the response is expected
            if response.status_code == 200:
                commits = response.json()
                
                for commit in commits:
                    try:
                        #Get the email, name and commit time from the commit object
                        sha = commit["sha"]
                        author_name = commit["commit"]["author"]["name"]
                        author_email = commit["commit"]["author"]["email"]
                        commit_time = commit["commit"]["author"]["date"]

                        data.append((sha, author_name, author_email, commit_time))
                    except:
                        continue
                
                if len(data) != 0:
                    #Store data into DB
                    store_into_database(owner, repo, data)
                    #Empty loaded data 
                    data = list()
                    
            #Calculate the url link for the next list of commits
            next_url = None
            link_header = response.headers.get("Link")
            if link_header:
                for link in link_header.split(", "):
                    if 'rel="next"' in link:
                        next_url = link[link.index("<") + 1: link.index(">")]
                        break
            
            url = next_url
            
    #Create the SQLite DB
    def create_database():
    
        conn = sqlite3.connect('github_commits.db')
        print("Database created!")

        conn.close()

    #Create the only table for the data we get from the API. Format of table: commits_{owner}_{repo}
    def create_table(owner, repo):

        conn = sqlite3.connect('github_commits.db')

        drop_query = get_secrets("DDL", "drop_commits_table").format(**{"owner": owner, "repo": repo})
        create_query = get_secrets("DDL", "create_commits_table").format(**{"owner": owner, "repo": repo})

        conn.execute(drop_query)
        conn.execute(create_query)

        conn.commit()
        conn.close()

    #Insert into the commits table
    def insert_into_table(owner, repo, data):

        conn = sqlite3.connect('github_commits.db')

        insert_query = get_secrets("DML", "insert_commits_table").format(**{"owner": owner, "repo": repo})

        conn.executemany(insert_query, data)

        conn.commit()
        conn.close()
        
    #Helper table to calculate the Weekday names for each day
    def create_weekday_table():

        conn = sqlite3.connect('github_commits.db')

        drop_query = get_secrets("DDL", "drop_weekday_table")
        create_query = get_secrets("DDL", "create_weekday_table")

        conn.execute(drop_query)
        conn.execute(create_query)

        conn.commit()
        conn.close()
        
    #Insert data into weekday table
    def insert_into_weekday_table():
        
        data = [(0, "0_Sunday"), (1, "1_Monday"), (2, "2_Tuesday"), (3, "3_Wednesday"), (4, "4_Thursday"), (5, "5_Friday"), (6, "6_Saturday")]
        
        conn = sqlite3.connect('github_commits.db')

        insert_query = get_secrets("DML", "insert_weekday_table")

        conn.executemany(insert_query, data)

        conn.commit()
        conn.close()
    
    #Helper table to calculate the time_of_day buckets for each hour
    def create_buckets_table():

        conn = sqlite3.connect('github_commits.db')
    
        drop_query = get_secrets("DDL", "drop_buckets_table")
        create_query = get_secrets("DDL", "create_buckets_table")

        conn.execute(drop_query)
        conn.execute(create_query)

        conn.commit()
        conn.close()
        
    #Insert data into buckets table
    def insert_into_buckets_table():
        
        data = [
            ("00", "02", "00-03"), 
            ("03", "05", "03-06"),
            ("06", "08", "06-09"),
            ("09", "11", "09-12"),
            ("12", "14", "12-15"),
            ("15", "17", "15-18"),
            ("18", "20", "18-21"),
            ("21", "23", "21-00")
        ]
        
        conn = sqlite3.connect('github_commits.db')

        insert_query = get_secrets("DML", "insert_buckets_table")

        conn.executemany(insert_query, data)

        conn.commit()
        conn.close()
    
    #Calls insert into table and prints the total rows loaded
    def store_into_database(owner, repo, data):
        insert_into_table(owner, repo, data)
        print("{length} records added to DB.".format(length = len(data)))
    
    db_loaded = False
    
    #Recursive loop to ask user for input and check if everthing is right. If so, fetch and load data into DB
    while not db_loaded:   
        owner, repo = ask_repo_url()
        start_date = ask_start_date()
        end_date = ask_end_date()
        
        #DB and tables
        prerequisites(owner, repo)
        
        #Check if the request is successful to the API
        if get_commits(owner, repo, start_date, end_date).status_code != 200:
            decision = input('Something went wrong!\n\nPlease make sure that the repo exists and the name and format are correct.\n\nCannot pull commits for private repositories.\n\nTry again? (y/n): ')
            
            if decision.lower() == 'n':
                print("Good day!")
                break
            else:
                continue
        else:
            get_all_commits(owner, repo, start_date, end_date)
            print("All records loaded into DB!")
            db_loaded = True
            
    return (owner, repo, db_loaded)

owner, repo, db_loaded = etl()
```
